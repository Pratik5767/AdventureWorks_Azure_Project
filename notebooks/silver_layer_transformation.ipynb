### Transformations

#### Calender

```python
display(df_cal)
```

```python
df_cal = (
    df_cal.withColumn('Date', to_date(col('Date'), 'M/d/yyyy'))
        .withColumn('Month', month(col('Date')))
        .withColumn('Year', year('Date'))
)

display(df_cal)
```

```python
df_cal.write.format("parquet") \
    .mode("append") \
    .option("path", "abfss://workspace_learn_100046@onelake.dfs.fabric.microsoft.com/adventureworks_LH.Lakehouse/Files/silver/AdventureWorks_Calendar") \
    .save()
```

#### Customers

```python
display(df_cus)
```

```python
df_cus = df_cus.withColumn('fullName', concat(col("Prefix"),lit(' '),col("FirstName"),lit(' '),col('LastName')))
display(df_cus)
```

```python
df_cus = df_cus.withColumn('fullName', concat_ws(' ', col('Prefix'), col('FirstName'), col('LastName')))
display(df_cus)
```

```python
df_cus.write.format("parquet") \
    .mode("append") \
    .option("path", "abfss://workspace_learn_100046@onelake.dfs.fabric.microsoft.com/adventureworks_LH.Lakehouse/Files/silver/AdventureWorks_Customers") \
    .save()
```

#### Sales

```python
display(df_sales)
```

```python
df_sales = df_sales.withColumn('StockDate', to_date(col('StockDate'), 'M/d/yyyy'))
```

```python
df_sales = df_sales.withColumn('StockDate', to_timestamp(col('StockDate')))
```

```python
df_sales = df_sales.withColumn('OrderNumber', regexp_replace(col('OrderNumber'), 'S', 'T'))
```

```python
df_sales = df_sales.withColumn('multiply', col('OrderLineItem') * col('OrderQuantity'))
```

```python
df_sales = df_sales.withColumn('OrderDate', to_date(col('OrderDate'), 'M/d/yyyy'))
```

```python
display(df_sales)
```

```python
df_sales.write.format("parquet") \
    .mode("append") \
    .option("path", "abfss://workspace_learn_100046@onelake.dfs.fabric.microsoft.com/adventureworks_LH.Lakehouse/Files/silver/AdventureWorks_Sales") \
    .save()
```
